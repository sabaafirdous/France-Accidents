# Trend Analysis

```python
# Extract relevant columns for temporal analysis
temporal_columns = ['year_of_acc', 'month_of_acc', 'day_of_acc', 'acc_tim_in_hour_min']
```

```python
# Create a custom function to parse the date and time
def parse_date_and_time(row):
    year = int(row['year_of_acc'])
    month = int(row['month_of_acc'])
    day = int(row['day_of_acc'])
    hour = row['acc_tim_in_hour_min'] // 100
    minute = row['acc_tim_in_hour_min'] % 100
    return pd.Timestamp(year, month, day, hour, minute)

# Apply the custom function to create a 'date' column
df['date'] = df.apply(parse_date_and_time, axis=1)
```

```python
# Plot temporal trends
plt.figure(figsize=(14, 14))

# Trend by year
plt.subplot(2, 2, 1)
yearly_trend = df.groupby('year_of_acc').size()
yearly_trend.plot(title='Accidents by Year', ylabel='Number of Accidents')

# Trend by month
plt.subplot(2, 2, 2)
monthly_trend = df.groupby('month_of_acc').size()
monthly_trend.plot(title='Accidents by Month', ylabel='Number of Accidents')

# Trend by day
plt.subplot(2, 2, 3)
daily_trend = df.groupby('day_of_acc').size()
daily_trend.plot(title='Accidents by Day', ylabel='Number of Accidents')

# Trend by time of day (hourly)
plt.subplot(2, 2, 4)
hourly_trend = df.groupby(df['date'].apply(lambda x: x.hour)).size()
hourly_trend.plot(title='Accidents by Hour of the Day', ylabel='Number of Accidents', xlabel='Hour of the Day')

plt.tight_layout()
plt.show()
```

```python
# Define the severity categories and their descriptions
severity_categories = [1, 2, 3, 4]
severity_dict = {
    1: 'Unscathed',
    2: 'Killed',
    3: 'Hospitalized wounded',
    4: 'Light injury'
}
```

```python
# Define the original lighting conditions values
lighting_conditions = [1, 2, 3, 4, 5]

# Create a mapping for the original lighting condition values to their descriptions
lighting_condition_mapping = {
    1: 'Full day',
    2: 'Twilight or dawn',
    3: 'Night without public lighting',
    4: 'Night with public lighting not lit',
    5: 'Night with public lighting on'
}

# Create a subplot with two columns to display charts side by side
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
fig.subplots_adjust(hspace=0.5)

# Create separate charts for each severity category
for i, severity in enumerate(severity_categories):
    # Filter the DataFrame to select rows with the specified severity category
    filtered_df = df[df['severity_of_the_accident'] == severity]

    # Initialize a list to store data for the current severity category
    severity_data = []

    # Loop through original lighting conditions values
    for condition in lighting_conditions:
        # Count the number of accidents for the current lighting condition in the filtered DataFrame
        count = len(filtered_df[filtered_df['lighting_condition'] == condition])

        # Append the count to the list
        severity_data.append(count)

    # Create a bar plot for the current severity category in the appropriate subplot
    ax = axs[i // 2, i % 2]
    sns.barplot(x=[lighting_condition_mapping[val] for val in lighting_conditions], y=severity_data, ax=ax)
    ax.set_title(f'Severity {severity_dict[severity]} Accidents by Lighting Conditions')
    ax.set_xlabel('Lighting Condition')
    ax.set_ylabel('Number of Accidents')
    ax.set_xticklabels([lighting_condition_mapping[val] for val in lighting_conditions], rotation=45)

    # Display the total number of accidents for the current severity condition
    total_accidents = sum(severity_data)
    ax.text(3, 0.9 * max(severity_data), f'Total: {total_accidents} accidents', fontsize=12, ha='center')

# Show the plots
plt.show()
```

# Trend Analysis

````python
# Extract relevant columns for temporal analysis
temporal_columns = ['year_of_acc', 'month_of_acc', 'day_of_acc', 'acc_tim_in_hour_min']
````

````python
# Create a custom function to parse the date and time
def parse_date_and_time(row):
    year = int(row['year_of_acc'])
    month = int(row['month_of_acc'])
    day = int(row['day_of_acc'])
    hour = row['acc_tim_in_hour_min'] // 100
    minute = row['acc_tim_in_hour_min'] % 100
    return pd.Timestamp(year, month, day, hour, minute)

# Apply the custom function to create a 'date' column
df['date'] = df.apply(parse_date_and_time, axis=1)
````

````python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a seaborn style for better aesthetics
sns.set_style("whitegrid")

# Create a figure with subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 14))

# Trend by year
yearly_trend = df.groupby('year_of_acc').size()
axs[0, 0].plot(yearly_trend.index, yearly_trend.values, marker='o', linestyle='-')
axs[0, 0].set_title('Accidents by Year', fontsize=14)
axs[0, 0].set_xlabel('Year', fontsize=12)
axs[0, 0].set_ylabel('Number of Accidents', fontsize=12)

# Trend by month
monthly_trend = df.groupby('month_of_acc').size()
axs[0, 1].plot(monthly_trend.index, monthly_trend.values, marker='o', linestyle='-')
axs[0, 1].set_title('Accidents by Month', fontsize=14)
axs[0, 1].set_xlabel('Month', fontsize=12)
axs[0, 1].set_ylabel('Number of Accidents', fontsize=12)

# Trend by day
daily_trend = df.groupby('day_of_acc').size()
axs[1, 0].plot(daily_trend.index, daily_trend.values, marker='o', linestyle='-')
axs[1, 0].set_title('Accidents by Day', fontsize=14)
axs[1, 0].set_xlabel('Day of the Month', fontsize=12)
axs[1, 0].set_ylabel('Number of Accidents', fontsize=12)

# Trend by time of day (hourly)
hourly_trend = df.groupby(df['date'].apply(lambda x: x.hour)).size()
axs[1, 1].plot(hourly_trend.index, hourly_trend.values, marker='o', linestyle='-')
axs[1, 1].set_title('Accidents by Hour of the Day', fontsize=14)
axs[1, 1].set_xlabel('Hour of the Day', fontsize=12)
axs[1, 1].set_ylabel('Number of Accidents', fontsize=12)

# Adjust spacing between subplots
plt.tight_layout()

# Show the plots
plt.show()
````

````python
# Define the severity categories and their descriptions
severity_categories = [1, 2, 3, 4]
severity_dict = {
    1: 'Unscathed',
    2: 'Killed',
    3: 'Hospitalized wounded',
    4: 'Light injury'
}
````

````python
# Define the original lighting conditions values
lighting_conditions = [1, 2, 3, 4, 5]

# Create a mapping for the original lighting condition values to their descriptions
lighting_condition_mapping = {
    1: 'Full day',
    2: 'Twilight or dawn',
    3: 'Night without public lighting',
    4: 'Night with public lighting not lit',
    5: 'Night with public lighting on'
}

import matplotlib.pyplot as plt
import seaborn as sns

# Create a subplot with two columns to display charts side by side
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
fig.subplots_adjust(hspace=0.5)

# Create a color palette for the bars
color_palette = sns.color_palette("Set2")

# Create separate charts for each severity category
for i, severity in enumerate(severity_categories):
    # Filter the DataFrame to select rows with the specified severity category
    filtered_df = df[df['severity_of_the_accident'] == severity]

    # Initialize a list to store data for the current severity category
    severity_data = []

    # Loop through original lighting conditions values
    for condition in lighting_conditions:
        # Count the number of accidents for the current lighting condition in the filtered DataFrame
        count = len(filtered_df[filtered_df['lighting_condition'] == condition])

        # Append the count to the list
        severity_data.append(count)

    # Create a bar plot for the current severity category in the appropriate subplot
    ax = axs[i // 2, i % 2]
    sns.barplot(x=[lighting_condition_mapping[val] for val in lighting_conditions], y=severity_data, ax=ax, palette=color_palette)
    ax.set_title(f'Severity {severity_dict[severity]} Accidents by Lighting Conditions', fontsize=14)
    ax.set_xlabel('Lighting Condition', fontsize=12)
    ax.set_ylabel('Number of Accidents', fontsize=12)
    ax.set_xticklabels([lighting_condition_mapping[val] for val in lighting_conditions], rotation=45, fontsize=10)

    # Display the total number of accidents for the current severity condition
    total_accidents = sum(severity_data)
    ax.text(0.5, 0.92 * max(severity_data), f'Total: {total_accidents} accidents', fontsize=12, ha='center', color='red')

# Show the plots
plt.tight_layout()
plt.show()
````
